# **Outlier Mitigation**

**IQR Method**

```python
#Check col types
df.dtypes
```

```python
def replace_outlier(df):
  for col in df.select_dtypes(include=['number']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5 * IQR
    ub = Q3 + 1.5 * IQR
    df[col] = df[col].apply(
        lambda x : Q1 if x < lb else (Q3 if x > ub else x)
    )
  return df
```

```python
df = replace_outlier(df)
```
